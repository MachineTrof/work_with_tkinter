# Крестики-Нолики </br>(feat. Полина Сементина и Михаил Трофимов)

```python
#Создаём холст 600х600
root = Tk()
root.title("Крестики-Нолики")
canvas = Canvas(root, width = 600, height = 600)
canvas.pack()

#Задаём нужные переменные: матрицу поля, за кем ход и счётчик ходов
Matrix = [[None, None, None], 
          [None, None, None], 
          [None, None, None]]
turn = 'X'
count_turn = 0

#Функция игры
def check_game(number):
    global turn, Matrix, count_turn
    
    #Переводим текстовые координаты в числовые, заполняем матрицу, рисуем фигуру
    n1 = int(number[0]) - 1
    n2 = int(number[1]) - 1
    if turn == 'X':
        Matrix[n1][n2] = 'X'
        canvas.create_polygon((200*n2 + 50, 200*n1 + 50), (200*n2 + 150, 200*n1 + 150), fill = 'black', outline = 'black')
        canvas.create_polygon((200*n2 + 50, 200*n1 + 150), (200*n2 + 150, 200*n1 + 50), fill = 'black', outline = 'black')
    else:
        Matrix[n1][n2] = 'O'
        canvas.create_oval(200*n2 + 50, 200*n1 + 50, 200*n2 + 150, 200*n1 + 150, fill = 'white', outline = 'orange')
    
    #Проверяем игру на конец игры
    end_game = False
    if (Matrix[0][0] == Matrix[0][1]) and (Matrix[0][1] == Matrix[0][2]) and (Matrix[0][1] != None):
        end_game = True
    if (Matrix[1][0] == Matrix[1][1]) and (Matrix[1][1] == Matrix[1][2]) and (Matrix[1][1] != None):
        end_game = True
    if (Matrix[2][0] == Matrix[2][1]) and (Matrix[2][1] == Matrix[2][2]) and (Matrix[2][1] != None):
        end_game = True
    if (Matrix[0][0] == Matrix[1][0]) and (Matrix[1][0] == Matrix[2][0]) and (Matrix[1][0] != None):
        end_game = True
    if (Matrix[0][1] == Matrix[1][1]) and (Matrix[1][1] == Matrix[2][1]) and (Matrix[1][1] != None):
        end_game = True
    if (Matrix[0][2] == Matrix[1][2]) and (Matrix[1][2] == Matrix[2][2]) and (Matrix[1][2] != None):
        end_game = True
    if (Matrix[0][0] == Matrix[1][1]) and (Matrix[1][1] == Matrix[2][2]) and (Matrix[1][1] != None):
        end_game = True
    if (Matrix[0][2] == Matrix[1][1]) and (Matrix[1][1] == Matrix[2][0]) and (Matrix[1][1] != None):
        end_game = True
    
    #Если игра завершилась - удаляем оставшиеся кнопки и определяем победителя
    if end_game == True:
        text.delete(1.0, 2.0)
        text.insert(1.0, f'Победил игрок "{turn}!"')
        text.tag_add("center", 1.0, "end")
        try: button11.destroy()
        except: pass
        try: button12.destroy()
        except: pass
        try: button13.destroy()
        except: pass
        try: button21.destroy()
        except: pass
        try: button22.destroy()
        except: pass
        try: button23.destroy()
        except: pass
        try: button31.destroy()
        except: pass
        try: button32.destroy()
        except: pass
        try: button33.destroy()
        except: pass
    
    #Иначе делаем проверку на ничью. Передаём ход
    else:
        if count_turn == 8:
            text.delete(1.0, 2.0)
            text.insert(1.0, 'Ничья!')
            text.tag_add("center", 1.0, "end")
        elif turn == 'X':
            turn = 'O'
            count_turn += 1
            text.delete(1.0, 2.0)
            text.insert(1.0, 'Ходит игрок "O"')
            text.tag_add("center", 1.0, "end")
        else:
            turn = 'X'
            count_turn += 1
            text.delete(1.0, 2.0)
            text.insert(1.0, 'Ходит игрок "X"')
            text.tag_add("center", 1.0, "end")

#Создаём кнопки с привязанными к ним функциями
def set_sign_11():
    check_game('11')
    button11.destroy()
cell11 = canvas.create_rectangle(0, 0, 200, 200, fill = 'white', outline = 'black')
button11 = Button(text = "", command = set_sign_11)
button11.place(x = 0, y = 0, width = 199, height = 199)

def set_sign_12():
    check_game('12')
    button12.destroy()
cell12 = canvas.create_rectangle(200, 0, 400, 200, fill = 'white', outline = 'black')
button12 = Button(text = "", command = set_sign_12)
button12.place(x = 201, y = 0, width = 199, height = 199)

def set_sign_13():
    check_game('13')
    button13.destroy()
cell13 = canvas.create_rectangle(400, 0, 600, 200, fill = 'white', outline = 'black')
button13 = Button(text = "", command = set_sign_13)
button13.place(x = 401, y = 0, width = 199, height = 199)

def set_sign_21():
    check_game('21')
    button21.destroy()
cell21 = canvas.create_rectangle(0, 200, 200, 400, fill = 'white', outline = 'black')
button21 = Button(text = "", command = set_sign_21)
button21.place(x = 0, y = 201, width = 199, height = 199)

def set_sign_22():
    check_game('22')
    button22.destroy()
cell22 = canvas.create_rectangle(200, 200, 400, 400, fill = 'white', outline = 'black')
button22 = Button(text = "", command = set_sign_22)
button22.place(x = 201, y = 201, width = 199, height = 199)

def set_sign_23():
    check_game('23')
    button23.destroy()
cell23 = canvas.create_rectangle(400, 200, 600, 400, fill = 'white', outline = 'black')
button23 = Button(text = "", command = set_sign_23)
button23.place(x = 401, y = 201, width = 199, height = 199)

def set_sign_31():
    check_game('31')
    button31.destroy()
cell31 = canvas.create_rectangle(0, 400, 200, 600, fill = 'white', outline = 'black')
button31 = Button(text = "", command = set_sign_31)
button31.place(x = 0, y = 401, width = 199, height = 199)

def set_sign_32():
    check_game('32')
    button32.destroy()
cell32 = canvas.create_rectangle(200, 400, 400, 600, fill = 'white', outline = 'black')
button32 = Button(text = "", command = set_sign_32)
button32.place(x = 201, y = 401, width = 199, height = 199)

def set_sign_33():
    check_game('33')
    button33.destroy()
cell33 = canvas.create_rectangle(400, 400, 600, 600, fill = 'white', outline = 'black')
button33 = Button(text = "", command = set_sign_33)
button33.place(x = 401, y = 401, width = 199, height = 199)

#Текстовый виджет снизу экрана
text = Text(bg = 'white', font = 'Arial 40', height = 1, width = 25)
text.tag_configure("center", justify='center')
text.pack()
text.insert(1.0, 'Ходит игрок "Х"')
text.tag_add("center", 1.0, "end")

#Цикл
root.mainloop()
```
