```python
from tkinter import *
from math import (sin, cos, pi)
from random import (randint, choice)
```

Необходимо написать двумерную аркадную игру «Астероиды» по следующим требованиям:
1. При старте игры на экране отображается (см. рисунки): </br>
a. Начальное количество жизней(3); </br>
b. Начальное количество очков (0); </br>
c. неподвижныйфон; </br>
d. динамический фон – полупрозрачное изображение астероидов, которое движется вдоль экрана влево с постоянной скоростью; </br>
e. заставкаигры–поцентру.
2. При щелчке на заставке игра начинается. В центре игровой области появляется корабль. Также начинают генерироваться астероиды в произвольных местах игрового поля с определенной частотой.
3. Корабль управляется игроком с клавиатуры и может:
a. вращатьсяотносительносвоегоцентра;
b. стрелять ракетами по направлению ориентации корабля; c. ускорятьсяпонаправлениюориентациикорабля;
4. Игровое поле имеет тороидальную геометрию, то есть при пересечении любым объектом, например, левой границы экрана он появляется с правой стороны, сохраняя направление и величину линейной и угловой скорости.
5. При ускорении изображение корабля должно меняться на изображение корабля с включенными двигателями. При завершении ускорения изображение должно меняться на исходное.
6. При выключенных двигателях корабля, он испытывает сопротивление среды и постепенно замедляется до полной остановки.
7. Астероиды при появлении имеют случайную линейную и угловую скорость.
8. Ракеты должны появляться из «носа» корабля и лететь в ту сторону, в которую ориентировано изображение корабля с постоянной скоростью.
9. Ракеты имеют ограниченное время жизни. По его истечении ракета исчезает с экрана. Это время жизни нужно выбрать исходя из отношения размеров экрана к скорости полета ракеты таким образом, чтобы ракета не могла пролететь все игровое поле даже по малому измерению.
10.При столкновении ракеты и астероида астероид и ракета исчезают с экрана, на месте астероида показывается анимация взрыва, игроку начисляется одно очко.
11.При столкновении корабля и астероида, астероид исчезает с экрана, на его месте показывается анимация взрыва, корабль возрождается неподвижный в середине игрового поля, количество жизней игрока уменьшается на единицу.
12.Если на момент смерти игрока, количество жизней равнялось 0, игра считается оконченной, на экране снова демонстрируется заставка (все астероиды, ракеты и корабль не прорисовываются), счет и количество жизней остаются на экране.
13.Астероиды не сталкиваются, а свободно проходят друг поверх друга.

Указания к выполнению лабораторной:
1. Все необходимые параметры игры выбираются самостоятельно исходя из интереса игры.
2. Код программы надлежит разбить на несколько (2 – 4) файлов исходя из логики и соображений читаемости. В частности, рекомендуется определения классов выносить в отдельные файлы и импортировать их в основную программу.
3. Следует обратить внимание, что при показе заставки начинает игру только щелчок мыши на прямоугольнике заставки. Щелчок по игровому полю вне заставки должен игнорироваться.
4. Рекомендуемая схема управления игрой: <br/>
a. Space-выстрелракетой; <br/>
b. стрелки влево, вправо – вращение корабля; <br/>
c. стрелкавверх–включениедвигателя(ускорение)корабля. <br/>
5. Все параметры игры должны быть оформлены как глобальные переменные в определенном месте программы и должным образом прокомментированы.
6. Перед началом написания кода программы следует провести объектно-ориентированное проектирование, разработать иерархию классов, выделить вспомогательные функции и функции-обработчики событий. Все
функции, классы и методы должны быть тщательно продокументированы. Также для них должны быть выбраны короткие, но понятные имена, поясняющие их суть.

```python
#Создаём холст 600х600
root = Tk()
root.title("Стероиды")
canvas = Canvas(root, width = 600, height = 600)
canvas.pack()

#Задаём начальные характеристики корабля
center_x_ship = 300
center_y_ship = 300
angle = 0
speed = 0

#Создаём объект самого корабля
SpaceShip = canvas.create_polygon((center_x_ship-15, center_y_ship+15), (center_x_ship+15, center_y_ship+15), 
                                  (center_x_ship, center_y_ship-15), fill = 'red', outline = 'black')

#Задаём иные необходимые переменные - массивы ракет и астероидов, флаг выхода + счётчики очков, жизней и неуязвимости
#(По поводу флага выхода: холст обновляется быстро, не хочется, чтобы несколько раз выводился конец игры)
Shots = []
Steroids = []
exit = False
score = 0
live = 3
time_of_unkill = 0

#Функция ускорения корабля
def move_ship(Up):
    global speed, live
    if live != 0:
        speed += 1
    if speed > 4:
        speed = min(4, speed)

#Функция поворота влево
def rotate_ship_L(Left):
    global angle, live
    if live != 0:
        angle -= 5

#Функция поворота вправо
def rotate_ship_R(Right):
    global angle, live
    if live != 0:
        angle += 5
    
#Функция рисования ракет и добавления их в массив (центр ракеты, угол поворота и время жизни)
def shot_ship(Shot):
    global center_x_ship, center_y_ship, angle, Shots, live
    if live != 0:
        sh = canvas.create_polygon((center_x_ship + 15*sin(pi/180*angle), center_y_ship - 15*cos(pi/180*angle)),
                                   (center_x_ship + 25*sin(pi/180*angle), center_y_ship - 25*cos(pi/180*angle)),
                                   fill = 'black', outline = 'black')
        Shots.append([sh, center_x_ship, center_y_ship, angle, 0])

#Функция хода игры
def game():
    #Импортируем все переменные как глобальные
    global Shots, Steroids, score, live, center_x_ship, center_y_ship, angle, time_of_unkill, exit, speed
    
    #Если корабль ещё жив - он должен двигаться!
    if live != 0:
        #Перемещаем центр корабля, если он вышел за границы поля - сдвигаем его
        center_x_ship += speed*sin(pi/180*angle)
        center_y_ship -= speed*cos(pi/180*angle)
        if center_x_ship < 0:
            center_x_ship = 600 + center_x_ship
        if center_x_ship > 600:
            center_x_ship = center_x_ship - 600
        if center_y_ship < 0:
            center_y_ship = 600 + center_y_ship
        if center_y_ship > 600:
            center_y_ship = center_y_ship - 600
        #Перерисовываем корабль и снижаем его скорость
        canvas.coords(SpaceShip, 
                      center_x_ship - 450**0.5*sin(pi/180*(angle-45)), 
                      center_y_ship + 450**0.5*cos(pi/180*(angle-45)), 
                      center_x_ship + 450**0.5*sin(pi/180*(angle-135)), 
                      center_y_ship - 450**0.5*cos(pi/180*(angle-135)), 
                      center_x_ship + 15*sin(pi/180*angle),
                      center_y_ship - 15*cos(pi/180*angle))
        if speed > 0:
            speed -= 0.02
    
    #Спавним Стероиды по краям поля, если некоторые из них оказались уничтожены
    if len(Steroids) != 5:
        while len(Steroids) != 5:
            #Массив Стероидов хранит следующую инфу об уникальном Стероиде: его центр, радиус и угол поворота
            x, y = choice([(0, randint(0,600)), (600, randint(0,600)),
                          (randint(0,600), 0), (randint(0,600), 600)])
            radius = randint(10,40)
            ang = randint(-180, 180)
            steroid = canvas.create_oval(x + radius, y + radius, x - radius, y - radius, 
                                         fill = 'yellow', outline = 'black')
            Steroids.append([steroid, x, y, radius, ang])
    
    #Рисуем Стероиды
    for Steroid in Steroids:
        Steroid[1] += 2*sin(pi/180*Steroid[4])
        Steroid[2] -= 2*cos(pi/180*Steroid[4])
        if Steroid[1] < 0:
            Steroid[1] = 600 + Steroid[1]
        if Steroid[1] > 600:
            Steroid[1] = Steroid[1] - 600
        if Steroid[2] < 0:
            Steroid[2] = 600 + Steroid[2]
        if Steroid[2] > 600:
            Steroid[2] = Steroid[2] - 600
        canvas.coords(Steroid[0], Steroid[1] + Steroid[3], Steroid[2] + Steroid[3],
                      Steroid[1] - Steroid[3], Steroid[2] - Steroid[3])
    
    #Если корабль стрелял - двигаем ракеты
    if len(Shots) != 0:
        for shot in Shots:
            shot[1] += 6*sin(pi/180*shot[3])
            shot[2] -= 6*cos(pi/180*shot[3])
            if shot[1] < 0:
                shot[1] = 600 + shot[1]
            if shot[1] > 600:
                shot[1] = shot[1] - 600
            if shot[2] < 0:
                shot[2] = 600 + shot[2]
            if shot[2] > 600:
                shot[2] = shot[2] - 600
            canvas.coords(shot[0], shot[1] + 15*sin(pi/180*shot[3]), shot[2] - 15*cos(pi/180*shot[3]),
                          shot[1] + 25*sin(pi/180*shot[3]), shot[2] - 25*cos(pi/180*shot[3]))
            shot[4] += 1
            #Если ракета достаточно долго просуществовала - она взрывается, не достигнув цели
            if shot[4] == 80:
                Shots.remove(shot)
                canvas.delete(shot[0])
    
    #Если ракета ударилась об Стероид - добавляем себе очки и удаляем объекты
    for steroid in Steroids:
        for shot in Shots:
            if ((steroid[1]-shot[1])**2 + (steroid[2]-shot[2])**2) <= steroid[3]**2:
                try:
                    Steroids.remove(steroid)
                    Shots.remove(shot)
                    canvas.delete(steroid[0])
                    canvas.delete(shot[0])
                    score += 1
                    scores.delete(1.0, 2.0)
                    scores.insert(1.0, f'Очки: {score}')
                    scores.tag_add("center", 1.0, "end")
                except:
                    pass
    
    #Если корабль врезался в Стероид - телепортируем его в центр, даём ему неуязвимость
    #Отнимаем жизнь, разрушаем Стероид
    for steroid in Steroids:
        damage = False
        if ((center_x_ship - 450**0.5*sin(pi/180*(angle-45)) - steroid[1])**2 + 
            (center_y_ship + 450**0.5*cos(pi/180*(angle-45)) - steroid[2])**2) <= steroid[3]**2:
            damage = True
        if ((center_x_ship + 450**0.5*sin(pi/180*(angle-135)) - steroid[1])**2 + 
            (center_y_ship - 450**0.5*cos(pi/180*(angle-135)) - steroid[2])**2) <= steroid[3]**2:
            damage = True
        if ((center_x_ship + 15*sin(pi/180*angle) - steroid[1])**2 + 
            (center_y_ship - 15*cos(pi/180*angle) - steroid[2])**2) <= steroid[3]**2:
            damage = True
        if damage == True and time_of_unkill == 0 and live > 0:
            live -= 1
            Steroids.remove(steroid)
            canvas.delete(steroid[0])
            lives.delete(1.0, 2.0)
            lives.insert(1.0, f'Жизни: {live}')
            lives.tag_add("center", 1.0, "end")
            time_of_unkill = 20000
            canvas.itemconfig(SpaceShip, fill = 'blue')
            center_x_ship = 300
            center_y_ship = 300
            angle = 0
            canvas.coords(SpaceShip, center_x_ship-15, center_y_ship+15, center_x_ship+15, center_y_ship+15,
                          center_x_ship, center_y_ship-15)
        if time_of_unkill > 0:
            time_of_unkill -= 20
            if time_of_unkill == 0:
                canvas.itemconfig(SpaceShip, fill = 'red')
        #Корабль потерял все жизни - завершаем игру!
        if live == 0 and exit == False:
            exit = True
            end = Text(bg = 'white', height = 1, font = 'Arial 40', width = 14)
            end.tag_configure("center", justify='center')
            end.pack()
            end.insert(1.0, 'Игра Окончена!')
            end.tag_add("center", 1.0, "end")
            canvas.itemconfig(SpaceShip, fill = 'red')
    root.after(20, game)

#Виджет подсчёта очков
scores = Text(bg = 'white', height = 1, font = 'Arial 20', width = 7)
scores.tag_configure("center", justify='center')
scores.place(x = 515, y = 0)
scores.insert(1.0, 'Очки: 0')
scores.tag_add("center", 1.0, "end")

#Виджет количества жизней
lives = Text(bg = 'white', height = 1, font = 'Arial 20', width = 8)
lives.tag_configure("center", justify='center')
lives.place(x = 0, y = 0)
lives.insert(1.0, 'Жизни: 3')
lives.tag_add("center", 1.0, "end")

#Виджет начала игры
Game_Go = Text(bg = 'white', height = 1, font = 'Arial 40', width = 12)
Game_Go.tag_configure("center", justify='center')
Game_Go.place(x = 170, y = 280)
Game_Go.insert(1.0, 'НАЧАТЬ ИГРУ')
Game_Go.tag_add("center", 1.0, "end")

#Функция начала игры
def game_go(GO):
    Game_Go.destroy()
    game()

#Биндим клавиши
root.bind('<Up>', move_ship)
root.bind('<Left>', rotate_ship_L)
root.bind('<Right>', rotate_ship_R)
root.bind('<space>', shot_ship)
Game_Go.bind('<Button-1>', game_go)

#Цикл безумия
root.mainloop()
```
